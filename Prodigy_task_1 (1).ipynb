```python
df = pd.read_csv(
    r"C:\Users\parna\population_data.csv\API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv",
    skiprows=4
)

df.head()
```

```python
population_2020 = df[['Country Name', '2020']].dropna()
population_2020.head()
```

```python
import matplotlib.pyplot as plt

top10 = population_2020.sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['Country Name'], top10['2020'])
plt.title("Top 10 Most Populated Countries (2020)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```
